각 답변자의 caseid를 집어 넣으면 응답자의 임신 결과를 리스트로 출력하는 dict 만들기

아래의 두 코드는 완전히 같은 코드입니다

```python
from collections import defaultdict
preg_map = defaultdict(list) #초기화.
for index, caseid in preg.caseid.iteritems(): #caseid라는 column이 있습니다. 
  preg_map[caseid].append(index) #pregmap 이라는 딕셔너리의 caseid의  key에 index값을 value에 넣습니다. 
  # 예를 들어 caseid=1인 key의 value는 0,1가 들어갑니다. 
  # 그래서 caseid =10229인 key의 value는 해당 caseid값이 10229인 인덱스들이 들어가는게 맞습니다.
  # pregmap이라는 딕셔녀리가 위에서 완성되면, 단순히 아래에서는 이 딕셔너리의 key값이 10229인 value값만 가지고 옵니다.
caseid = 10229
indices = preg_map[caseid]
#indices # [11093, 11094, 11095, 11096, 11097, 11098, 11099] 가 나옴. 각각의 응답자들의 넘버입니다. 
preg.outcome[indices].values  # 위의 인덱스들에 해당하는 outcome 출력입니다!!!!
# preg의 outcome 칼럼의 원소들을 출력하고 싶은데 특정 인덱스의 것만 출력하겠다.
```

```python
#위의 과정을 nsfg.MakePregMap함수로 구현해놓았음. 
caseid = 10229
preg_map = nsfg.MakePregMap(preg)
indices = preg_map[caseid] #안에 멀 넣든 주어진 조건을 만족하는 인덱스들이 나옴.
preg.outcome[indices].values 
```
